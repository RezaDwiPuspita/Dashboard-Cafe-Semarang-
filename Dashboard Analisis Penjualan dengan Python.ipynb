```python
# Data preparation
data['Items'] = data['Items'].apply(lambda item: item.lower())
data['Items'] = data['Items'].apply(lambda item: item.strip())

data = data[['Gross Sales', 'Items']].copy()

data.head(10)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Menghitung jumlah item berdasarkan Items dan Gross Sales
items_count = data.groupby(['Items', 'Gross Sales'])['Items'].count().reset_index(name='Count')
print(items_count.head(10))


# Pivot table
items_count_pivot = items_count.pivot_table(
    index='Gross Sales',
    columns='Items',
    values='Count',
    aggfunc='sum'
).fillna(0)

print("Ukuran Dataset:", items_count_pivot.shape)
print(items_count_pivot.head(10))
```

```python
# Mengubah tipe data
items_count_pivot = items_count_pivot.astype('int32')
print(items_count_pivot.head())
```

```python
# Fungsi encoding
def encode(x):
    if x <= 0:
        return 0
    elif x >= 1:
        return 1


# Menerapkan fungsi encoding
items_count_pivot = items_count_pivot.applymap(encode)
print(items_count_pivot.head(10))
```

```python
# Menampilkan ukuran dataset dan jumlah Gross Sales dan Items
print('Ukuran dataset:', items_count_pivot.shape)
print('Jumlah Gross Sales:', items_count_pivot.shape[0])
print('Jumlah Items:', items_count_pivot.shape[1])
```

```python
# Menggunakan fungsi apriori setelah konversi tipe data
support = 0.05
frequent_items = apriori(items_count_pivot, min_support=support, use_colnames=True)
```

```python
# Mengurutkan dan menampilkan itemsets yang sering muncul
frequent_items_sorted = frequent_items.sort_values('support', ascending=False).head(10)
print(frequent_items_sorted)
```

```python
# Membuat aturan asosiasi
metric = "lift"
min_threshold = 1.0

rules = association_rules(
    frequent_items,
    metric=metric,
    min_threshold=min_threshold
)[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

rules.sort_values('lift', ascending=False, inplace=True)
rules.head(10)
```
